```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Snow.xlsx' , sheet_name= 'Sheet1')
df
```

```python
pd.set_option('display.max_colwidth' , None)
```

```python
df['Info'] = df['Info'].fillna('')
df
```

```python
# split the text column by return character and convert to a Series
split_text = df['Info'].apply(lambda x: pd.Series(x.split('\n')))
```

```python
split_text
```

```python
split_text.columns = ['Name', 'NA', 'Company' , 'Role']
split_text
```

```python
Final_df = split_text['Name'].apply(lambda x : pd.Series(x.split(' ')))
```

```python
ndf = Final_df[[0,1]]
ndf.columns = ['First Name' , 'Last Name']
```

```python
ndf
```

```python
fdf = ndf.join(split_text[['Company' , 'Role']])
```

```python
fdf
```

```python
rdf = fdf.join(df['EMAILS']) 
```

```python
rdf = rdf[rdf['EMAILS'] != "Can't find emails"] # dropping columns with can't find emails
```

```python
rdf = rdf.sort_values(by = 'Company' , ignore_index=True) # Ignore_index parameter is used here to keep index as it is , 
# otherwise it gets messed up as it sticks to the row values
```

```python
# sort the DataFrame by column 'A'
df_sorted = rdf

# create a new column 'partition' using cumcount
df_sorted['Rank'] = (df_sorted.groupby('Company').cumcount() + 1).astype(int)

df_sorted.loc[:20]
```

```python
df_sorted.to_excel('Finishedd.xlsx', index=False, sheet_name='Sheet1')
```
